```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import matplotlib.colors as colors


cit_astro = []
uni_astro = []
cit_phy = []
uni_phy = []

cit_bro = 0
cit_col = 0
cit_cor = 0
cit_dar = 0
cit_har = 0
cit_pen = 0
cit_pri = 0
cit_yal = 0

affil = ['Brown University', 'Columbia University', 'Cornell University', 'Dartmouth College', 'Harvard University', 'University of Pennsylvania', 'Princeton University', 'Yale University'] ### only look at Ivy League

f1 = open('Uni_astro.txt').readlines()[1:] ## open file with citations for astrophysics
for line in f1:
    l = line.split()
    cit_astro.append(int(l[0]))
    uni_astro.append(l[1]+l[2])

f2 = open('Uni_phy.txt').readlines()[1:] ## open file with citations for physics (without astro)
for line in f2:
    l = line.split()
    cit_phy.append(int(l[0]))
    uni_phy.append(l[1]+l[2])

#### total citations in astr for each University ####
for i in range(len(cit_astro)):
    if uni_astro[i] == '"BrownUniversity"':
        cit_bro += cit_astro[i]
    elif uni_astro[i] == '"ColumbiaUniversity"':
        cit_col += cit_astro[i]
    elif uni_astro[i] == '"CornellUniversity"':
        cit_cor += cit_astro[i]
    elif uni_astro[i] == '"DartmouthCollege"':
        cit_dar += cit_astro[i]
    elif uni_astro[i] == '"HarvardUniversity"':
        cit_har += cit_astro[i]
    elif uni_astro[i] == '"PrincetonUniversity"':
        cit_pri += cit_astro[i]
    elif uni_astro[i] == '"YaleUniversity"':
        cit_yal += cit_astro[i]
    else:
        cit_pen += cit_astro[i]

#### total citations in physics for each University ####

for i in range(len(cit_phy)):
    if uni_phy[i] == '"BrownUniversity"':
        cit_bro += cit_phy[i]
    elif uni_phy[i] == '"ColumbiaUniversity"':
        cit_col += cit_phy[i]
    elif uni_phy[i] == '"CornellUniversity"':
        cit_cor += cit_phy[i]
    elif uni_phy[i] == '"DartmouthCollege"':
        cit_dar += cit_phy[i]
    elif uni_phy[i] == '"HarvardUniversity"':
        cit_har += cit_phy[i]
    elif uni_phy[i] == '"PrincetonUniversity"':
        cit_pri += cit_phy[i]
    elif uni_phy[i] == '"YaleUniversity"':
        cit_yal += cit_phy[i]
    else:
        cit_pen += cit_phy[i]
        
cite = [cit_bro, cit_col, cit_cor, cit_dar, cit_har, cit_pen, cit_pri, cit_yal]        
staff = [30, 47, 53, 40, 76, 49, 72, 63] ## total proffessors in the physics department in each University
students = [83, 110, 167, 36, 220, 107, 130, 118] ## total Ph.D. students in the physics department in each University

colors = cm.rainbow(np.linspace(0, 1, len(cite)))     # create a bunch of colors

plt.suptitle("Citation vs. Size of department")
ax1 = plt.subplot(211)
ax1.set_xlabel("Number of Faculty")
ax1.set_ylabel("Citations")

ax2 = plt.subplot(212)
ax2.set_xlabel("Number of Graduate Students")
ax2.set_ylabel("Citations")


for i,c in enumerate(colors):

    x = staff[i]
    y = cite[i]
    l = affil[i]

    x2 = students[i]
    ax1.scatter(x, y, label=l, s=50, linewidth=0.1, c=c)
    ax2.scatter(x2, y, label=l, s=50, linewidth=0.1, c=c)
    

ax1.legend(bbox_to_anchor=(1.7, 0.5))


plt.tight_layout()
plt.show()


```

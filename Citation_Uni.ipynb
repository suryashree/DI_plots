```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

cit_astro = []
uni_astro = []
cit_phy = []
uni_phy = []

cit_bro = 0
cit_col = 0
cit_cor = 0
cit_dar = 0
cit_har = 0
cit_pen = 0
cit_pri = 0
cit_yal = 0

cit_brop = 0
cit_colp = 0
cit_corp = 0
cit_darp = 0
cit_harp = 0
cit_penp = 0
cit_prip = 0
cit_yalp = 0

affil = ['Brown University', 'Columbia University', 'Cornell University', 'Dartmouth College', 'Harvard University', 'University of Pennsylvania', 'Princeton University', 'Yale University'] ### only look at Ivy League

f1 = open('Uni_astro.txt').readlines()[1:] ## open file with citations for astrophysics
for line in f1:
    l = line.split()
    cit_astro.append(int(l[0]))
    uni_astro.append(l[1]+l[2])

f2 = open('Uni_phy.txt').readlines()[1:] ## open file with citations for physics (without astro)
for line in f2:
    l = line.split()
    cit_phy.append(int(l[0]))
    uni_phy.append(l[1]+l[2])

#### total citations in astr for each University ####
for i in range(len(cit_astro)):
    if uni_astro[i] == '"BrownUniversity"':
        cit_bro += cit_astro[i]
    elif uni_astro[i] == '"ColumbiaUniversity"':
        cit_col += cit_astro[i]
    elif uni_astro[i] == '"CornellUniversity"':
        cit_cor += cit_astro[i]
    elif uni_astro[i] == '"DartmouthCollege"':
        cit_dar += cit_astro[i]
    elif uni_astro[i] == '"HarvardUniversity"':
        cit_har += cit_astro[i]
    elif uni_astro[i] == '"PrincetonUniversity"':
        cit_pri += cit_astro[i]
    elif uni_astro[i] == '"YaleUniversity"':
        cit_yal += cit_astro[i]
    else:
        cit_pen += cit_astro[i]

#### total citations in physics for each University ####

for i in range(len(cit_phy)):
    if uni_phy[i] == '"BrownUniversity"':
        cit_brop += cit_phy[i]
    elif uni_phy[i] == '"ColumbiaUniversity"':
        cit_colp += cit_phy[i]
    elif uni_phy[i] == '"CornellUniversity"':
        cit_corp += cit_phy[i]
    elif uni_phy[i] == '"DartmouthCollege"':
        cit_darp += cit_phy[i]
    elif uni_phy[i] == '"HarvardUniversity"':
        cit_harp += cit_phy[i]
    elif uni_phy[i] == '"PrincetonUniversity"':
        cit_prip += cit_phy[i]
    elif uni_phy[i] == '"YaleUniversity"':
        cit_yalp += cit_phy[i]
    else:
        cit_penp += cit_phy[i]

#### plot bar of citatations per University ####

index = np.arange(len(affil))*3

bar_width = 1.0

y_pos1 = [cit_brop, cit_colp, cit_corp, cit_darp, cit_harp, cit_penp, cit_prip, cit_yalp]
y_pos2 = [cit_bro, cit_col, cit_cor, cit_dar, cit_har, cit_pen, cit_pri, cit_yal]


plt.barh(index, y_pos1, 1.0, align='center', alpha=0.5, color = 'b', label = 'Physics')
plt.barh(index+bar_width, y_pos2, 1.0, align='center', alpha=0.5, color = 'r', label = 'Astronomy & Astrophysics')


plt.yticks(index+(bar_width/2.), affil)
plt.xlabel('Citations')
plt.title('Citations over the period 2013 - 2018')
plt.legend(loc='best')

plt.show()


```
